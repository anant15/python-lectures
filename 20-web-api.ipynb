```python
# API is the two computers/services interact with each other, when GUI is not needed

# GET requests and post requests
import requests
response = requests.get("https://www.google.com")
```

```python
print(response.status_code) 

# 200-299 - Successfully
# 300-399 - Request was valid, but it couldn't be fulfilled
# 400-499 - Invalid requests
```

```python
print(response.headers)
```

```python
print(response.content)
```
